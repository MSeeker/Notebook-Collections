```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn
from collections import deque
%matplotlib inline
matplotlib.rc('figure', figsize=(8, 6))
np.random.seed(2015)
```

# 1. Ising模型回顾

N个相互作用的“分子”（磁偶极子），自旋量$\sigma_i = \pm 1$，总能量

$$ E=-J\sum_{ [i,j] }\sigma_i\sigma_j-H\sum_i\sigma_i $$

方格状的二维Ising模型：Onsager给出严格解

* 转变温度： $$ T_C = \frac{1}{k_B}\frac{2J}{\ln(1+\sqrt{2})} \approx 2.269\frac{J}{k_B} $$

* 磁化率（约化后）：$$ M \equiv |\bar{\sigma}| = \Big\{
    \begin{array}
    ((1 - \sinh^{-4}(\frac{2J}{k_B T}))^{1/8} & (T < T_C) \\
    0 & (T \geq T_C)
    \end{array}
    $$

* 二维Ising模型作为算法的“试金石”

# 2. 蒙特卡洛模拟

### 热力学系统模拟的核心思想：**Importance sampling**

* 即使对于粒子数$N \sim 10^4$的小系统，总微观状态数$\Omega$也是天文数字。

* 当$N$足够大时，除最概然分布附近的一小片微观态外，大部分状态的概率都是极低的。

* Metropolis：
> ...instead of choosing configurations randomly, then weighting them with $\exp(-E/k_BT)$, we choose configurations with a probability $\exp(-E/k_BT)$ and weight them evenly.

### 生成样本的方法：

* **分子动力学模拟（Molecular dynamics simulation，MD）**：从第一性原理出发，取系统真实的相轨上各点为样本。

* **蒙特卡洛模拟（Monte Carlo simulation，MC）**：通过相空间上的随机行走生成样本，通常藉由马可夫链的构造实现，是为**马可夫链蒙特卡洛（Markov chain Monte Carlo，MCMC）**。

## 2.1 Metropolis-Hastings算法

### 马可夫链的有关知识：

* 演化规律：$$ Pr[x_{n+1}=j|x_n=i] = t_{ij} (n=0,1,2,\ldots,\Omega) $$ 
$t_{ij}$
：**转移概率**，
$T=(t_{ij})$：
**转移矩阵**。

* **稳态分布（Stationary distribution）**：$$ \pi_j = \sum_{i=1}^{\Omega}t_{ij}\pi_i(j = 1, 2, \ldots, \Omega) $$ 马可夫链收敛到唯一稳态分布的条件：**各态遍历性（Ergocity）**

* **细致平衡条件（Detailed balance）**：对所有的$i$与$j$，$t_{ij}\pi_i = t_{ji}\pi_j$。满足细致平衡条件的分布均为稳态分布。

### 算法纲要

假设当前系统处于状态$a$：

* 按事先选择的**先验概率分布**$A(a \rightarrow b)$选取状态b.

* 依概率 $$ p(a \rightarrow b) = \min[1,\frac{\pi(b)}{\pi(a)}\frac{A(b \rightarrow a)}{A(a \rightarrow b)}] $$ 决定是否保留b作为下一步的状态，不然舍弃b，仍取a为系统状态。

由此构造出的马可夫链具有转移概率$t(a \rightarrow b) = A(a \rightarrow b)p(a \rightarrow b)$，显然细致平衡条件$\pi(a)t(a \rightarrow b) = \pi(b)t(b \rightarrow a)$成立。

$\pi(a)$可以是未约化的：对正则分布，取$\pi(a)=\exp(-E(a)/k_BT)$即可。

## 2.2 M-H算法示例

* 情景：一个灌了铅的骰子，各面朝上的概率为$\pi(1)=1/2$，$\pi(2)=1/4$，$\pi(3)=1/8$，$\pi(4)=1/16$，$\pi(5)=\pi(6)=1/32$
* 先验概率选取：均匀分布（$A(i \rightarrow j) = 1/6$）

```python
def dice_samples(trials):
    '''用Metropolis-Hastings算法模拟不均匀的骰子'''
    prob = {1: 1/2, 2: 1/4, 3: 1/8, 4: 1/16, 5: 1/32, 6: 1/32}
    samples = np.zeros(trials + 1, dtype=int)
    samples[0] = 1
    for i in range(trials):
        a = samples[i]
        b = np.random.random_integers(1, 6) # 均匀先验概率分布
        pa = prob[a]
        pb = prob[b]
        if pb >= pa or np.random.rand() < pb / pa:
            samples[i + 1] = b
        else:
            samples[i + 1] = a
    return samples
```

```python
def summarize(samples):
    '''统计各面朝上次数的占比'''
    num_samples = len(samples)
    distribution = {i: (samples == i).sum() * 100 / num_samples for i in [1, 2, 3, 4, 5 ,6]} # 百分数
    return distribution
```

```python
samples = dice_samples(1000000)
ns = np.array(np.logspace(1, 6, num=50), dtype=int)

# 统计不同步数下各面朝上次数的占比
distributions = {i: np.zeros(50) for i in [1, 2, 3, 4 ,5 ,6]}
for index in range(50):
    n = ns[index]
    distribution = summarize(samples[:n])
    for i in [1, 2, 3, 4, 5, 6]:
        distributions[i][index] = distribution[i]
for i in [1, 2, 3, 4, 5, 6]:
    plt.plot(ns, distributions[i], label='Face {}'.format(i))
plt.xlabel('Iteration counts')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.semilogx()
plt.legend()
```

# 3. M-H算法在Ising模型上的应用

Ising模型作为“邻居关系”的无向图：“邻居表”表示

```python
# 二维晶格中分子角标与其坐标之间的相互转换（a：边长）
def flatten_2d(i, j, a):
    return i * a + j
def unflatten_2d(n, a):
    j = n % a
    i = (n - j) // a
    return i, j

# 生成“邻居表”
def gen_neighbors_1d(N):
    '''生成长度为N的一维周期性边界条件晶链'''
    neighbors = np.zeros((N, 2), dtype=int)
    for n in range(N):
        neighbors[n][0] = (n - 1) % N # left
        neighbors[n][1] = (n + 1) % N # right
    return neighbors
def gen_neighbors_2d(a):
    '''生成a*a的二维周期性边界条件晶阵'''
    neighbors = np.zeros((a*a, 4), dtype=int)
    for n in range(a*a):
        i, j = unflatten_2d(n, a)
        neighbors[n][0] = flatten_2d(i, (j - 1) % a, a) # left
        neighbors[n][1] = flatten_2d(i, (j + 1) % a, a) # right
        neighbors[n][2] = flatten_2d((i - 1) % a, j, a) # up
        neighbors[n][3] = flatten_2d((i + 1) % a, j, a) # down
    return neighbors
```

先验概率分布：只考虑由目前的状态**翻转一个分子自旋**后可得到的$N$种微观态，并从中随机均匀地进行选择。

翻转前后能量差$\delta E= 2J\sigma_i\sum_{[i,j]}\sigma_j $，
接受翻转概率$ P=\min(1, \exp(-\delta E/k_BT)) $。

取$J=1$，$k_B=1$：

```python
def metropolis_hastings(neighbors_list, T, iterations):
    # 初始化
    size = neighbors_list.shape[0]
    spins = np.random.random_integers(0, 1, size)
    spins[spins == 0] = -1
    # 记录磁化率与总内能
    magnetization = np.zeros(iterations + 1)
    energy = np.zeros(iterations + 1)
    magnetization[0] = spins.sum()
    energy[0] = -spins.dot(spins[neighbors_list].sum(axis=1)) / 2
    
    for step in range(iterations):
        n = np.random.randint(0, size) # 随机选择分子，试着翻转之
        delta_E = 2 * spins[n] * spins[neighbors_list[n]].sum()
        if delta_E < 0 or np.random.rand() < np.exp(-delta_E / T):
            # 翻转自旋
            spins[n] = -spins[n]
            magnetization[step + 1] = magnetization[step] + 2 * spins[n]
            energy[step + 1] = energy[step] + delta_E
        else:
            # 保持原样
            magnetization[step + 1] = magnetization[step]
            energy[step + 1] = energy[step]
    return magnetization / size, energy
```

一维与二维Ising模型的结果：

```python
def plot_magnetization(dimension):
    if dimension == 1:
        neighbors_list = gen_neighbors_1d(400)
    elif dimension == 2:
        neighbors_list = gen_neighbors_2d(20)
    T_list = [0.5, 1.0, 1.5, 1.8, 2.0, 2.2, 2.4, 3.0, 3.5]
    fig = plt.figure(figsize=(12, 40))
    for i in range(9):
        T = T_list[i]
        magnetization, _ = metropolis_hastings(neighbors_list, T, 100000)
        # 随机行走历史
        fig.add_subplot(9, 2, 2 * i + 1)
        plt.plot(magnetization)
        plt.ylim(-1, 1)
        plt.ylabel('Magnetization')
        plt.xlabel('Iterations')
        plt.annotate('T = {}'.format(T), (10000,0.8))
        # 样本分布直方图
        fig.add_subplot(9, 2, 2 * i + 2)
        plt.hist(magnetization, bins=np.linspace(-1, 1, num=20), orientation='horizontal')
        plt.ylim(-1, 1)
        plt.xlabel('Counts')
```

```python
# 一维模拟
plot_magnetization(1)
```

```python
# 二维模拟
plot_magnetization(2)
```

评述：
* 一维的情况基本符合Ising的理论预言。
* 二维的结果在转变温度$T_C = 2.269$前后有着本质的不同，然而$T < T_C$时样本并未在两个稳态间均匀分布。
* 在$T_C$附近马可夫链很难收敛，可以证明$T \rightarrow T_C$时收敛所需步数是发散的，是为**临界慢化现象（Critical slowing down）**。

## 3.1 临界点附近自旋的集簇现象

定义自旋关联函数

$$ g(x, y) \equiv E[\sigma(m + x, n + y)\sigma(m, n)] - E[\sigma(m + x, m + y)]E[\sigma(m, n)] $$

其中所有的期望均是对格点$(m,n)$的平均。在周期性边界条件下，可化简为：

$$ g(x, y) = E[\sigma(m + x, n + y)\sigma(m, n)] - M^2 $$

其中M为约化的磁化率。以$T=2.2$为例考察集簇现象：

```python
def spin_correlation(spins, a):
    M = spins.mean()
    spins_2d = spins.reshape((a, a))
    rs = np.arange(-a/2, a/2 + 1, dtype=int)
    num_rs = len(rs)
    correlations = np.zeros((num_rs, num_rs))
    for i, y in enumerate(rs):
        for j, x in enumerate(rs):
            correlations[i, j] = (spins_2d * np.roll(np.roll(spins_2d, x, axis=1), y, axis=0)).mean() + M**2
    return correlations
```

```python
def gen_snapshots(time_stamps, a, T):
    iterations = time_stamps[-1]
    snapshots = {}
    
    # 与Metropolis-Hastings算法部分相同
    neighbors_list = gen_neighbors_2d(a)
    size = neighbors_list.shape[0]
    spins = np.random.random_integers(0, 1, size)
    spins[spins == 0] = -1
    for step in range(iterations):
        n = np.random.randint(0, size)
        delta_E = 2 * spins[n] * spins[neighbors_list[n]].sum()
        if delta_E < 0 or np.random.rand() < np.exp(-delta_E / T):
            spins[n] = -spins[n]
        if step + 1 in time_stamps:
            snapshots[step + 1] = {
                'spins': spins.copy().reshape((a, a)), 
                'magnetization': spins.mean(),
                'correlation': spin_correlation(spins, a)
            }
    return snapshots
```

```python
a = 40
T = 2.2
time_stamps = np.array([1, 10, 100, 1000, 10000, 20000, 40000, 60000, 80000, 100000])
snapshots = gen_snapshots(time_stamps, a, T)
with seaborn.axes_style('white'):
    fig, axes = plt.subplots(10, 2, figsize=(12, 6 * 10))
    for i, t in enumerate(time_stamps):
        axes[i][0].matshow(snapshots[t]['spins'], interpolation='none')
        axes[i][0].set_ylabel('T = {}, M = {:.3f}'.format(t, snapshots[t]['magnetization']))
        axes[i][0].set_xlabel('Spins')
        axes[i][1].matshow(snapshots[t]['correlation'])
        axes[i][1].set_xlabel('Spin correlation function')
        axes[i][1].set_xticks([0, 10, 20, 30, 40])
        axes[i][1].set_xticklabels([-20, -10, 0, 10, 20])
        axes[i][1].set_yticks([0, 10, 20, 30, 40])
        axes[i][1].set_yticklabels([-20, -10, 0, 10, 20])
```

集簇现象的出现大大降低了单分子自旋翻转的接受率，是临界慢化现象的一大原因。为此我们需要改变先验概率分布的形式，把随机行走的“步长”拉大。

## 3.2 Cluster algorithm

核心思想：翻转单个分子 → 翻转一大块具有相同自旋的cluster

算法步骤：
* 随机选取某个分子作为cluster构建的“种子”。
* 以该“种子”为根，在晶格中做广度优先搜索——如果碰到相同自旋的分子，则按给定的概率p将其添加到广度优先搜索树中，直至广搜结束。
* 依概率$p(a \rightarrow b)$决定是否将构建的cluster整体翻转。

假设cluster有$m$个分子、$n_1$段异号边界与$n_2$段同号边界，则先验概率

$$A(a \rightarrow b)=\frac{m}{N}p^{m-1}(1-p)^{n_2}$$

* $\frac{m}{N}$：“种子”选在cluster内部的概率。
* $p^{m-1}$：吸纳除“种子”外$m-1$个同号分子的概率。
* $(1-p)^{n_2}$：拒绝$n_2$个同号分子的概率。

考虑逆过程$b \rightarrow a$，同理（此时同号边界有$n_1$段，异号边界有$n_2$段）：

$$A(b \rightarrow a) = \frac{m}{N}p^{m-1}(1-p)^{n_1}$$

翻转前后只有边界上的分子对能量会有变化：

$$ \frac{\pi(b)}{\pi(a)} = e^{-\delta E(a \rightarrow b)/k_B T} = e^{-2(n_2-n_1)/k_BT}$$

取$p(a \rightarrow b) = \min[1, e^{-2(n_2-n_1)/k_BT}(1-p)^{n_1-n_2}]$容易验证这样的选择满足细致平衡条件 $\pi(a)A(a \rightarrow b)p(a \rightarrow b) = \pi(b)A(b \rightarrow a)p(b \rightarrow a)$。

事实上，若取$p=1-e^{2/k_BT}$，则$p(a \rightarrow b) \equiv 1$。在这种参数选取下，每一步随机行走都是有效的，这大大提高了马可夫链的收敛速度。

```python
def cluster_MC(neighbors_list, T, iterations):
    p = 1 - np.exp(-2 / T) # 理想的p值
    # 初始化
    size = neighbors_list.shape[0]
    spins = np.random.random_integers(0, 1, size)
    spins[spins == 0] = -1
    # 记录磁化率与能量
    magnetization = np.zeros(iterations + 1)
    magnetization[0] = spins.sum()
    energy = np.zeros(iterations + 1)
    energy[0] = -spins.dot(spins[neighbors_list].sum(axis=1))
    
    for step in range(iterations):
        # 广搜构建cluster（借助队列）
        n0 = np.random.randint(0, size)
        sign = spins[n0]
        cluster = set([n0])
        pockets = deque([n0])
        finished = False
        while not finished:
            try:
                n = pockets.popleft()
                neighbors = neighbors_list[n]
                for neighbor in neighbors:
                    if spins[neighbor] == sign and neighbor not in cluster and np.random.rand() < p:
                        cluster.add(neighbor)
                        pockets.append(neighbor)
            except IndexError: # 广搜完毕
                finished = True
        # 翻转cluster
        cluster = np.fromiter(cluster, dtype=int)
        spins[cluster] = -sign
        magnetization[step + 1] = magnetization[step] - 2 * sign * len(cluster)
        energy[step + 1] = -spins.dot(spins[neighbors_list].sum(axis=1))
    return magnetization / size, energy / 2 # 重复计数
```

让我们与单分子翻转的算法做个对比：

```python
def plot_magnetization_cluster(dimension):
    if dimension == 1:
        neighbors_list = gen_neighbors_1d(400)
    elif dimension == 2:
        neighbors_list = gen_neighbors_2d(20)
    T_list = [0.5, 1.0, 1.5, 1.8, 2.0, 2.2, 2.4, 3.0, 3.5]
    fig = plt.figure(figsize=(12, 40))
    for i in range(9):
        T = T_list[i]
        magnetization, _ = cluster_MC(neighbors_list, T, 1000)
        # 随机行走历史
        fig.add_subplot(9, 2, 2 * i + 1)
        plt.plot(magnetization)
        plt.ylim(-1, 1)
        plt.ylabel('Magnetization')
        plt.xlabel('Iterations')
        plt.annotate('T = {}'.format(T), (50,0.8))
        # 样本分布直方图
        fig.add_subplot(9, 2, 2 * i + 2)
        plt.hist(magnetization, bins=np.linspace(-1, 1, num=20), orientation='horizontal')
        plt.ylim(-1, 1)
        plt.xlabel('Counts')
```

```python
plot_magnetization_cluster(2)
```

## 3.3 模拟结果

我们采取cluster算法。磁化率$M$、内能$E$可以直接系综平均而得，热容$C_V$略经变形后也能表示为样本的统计值：

$$ C_V = \frac{\partial\bar{E}}{\partial T} = \frac{1}{k_B T}(\bar{E^2} - {\bar{E}}^2) = \frac{1}{k_BT}\sigma_E^2$$

```python
def averages(magnetization, energy, T):
    '''计算各量的系综平均'''
    M = np.abs(magnetization).mean()
    E = energy.mean()
    CV = energy.var() / T
    return M, E, CV
```

```python
points = 31
dimension = 64
iterations = 5000

neighbors_list = gen_neighbors_2d(dimension)
Ts = np.linspace(1.0, 4.0, points)
Ms = np.zeros(points)
Es = np.zeros(points)
CVs = np.zeros(points)
for i in range(points):
    T = Ts[i]
    magnetization, energy = cluster_MC(neighbors_list, T, iterations)
    Ms[i], Es[i], CVs[i] = averages(magnetization, energy, T)
    print("T = {:.4f}, M = {:.4f}, E = {:.4f}, CV = {:.4f}".format(T, Ms[i], Es[i], CVs[i]))
```

```python
# 磁化率
Onsager_Tc = 2.269
Ts_plot = np.linspace(1.0, 4.0, num=50)
Onsager_Ms = np.zeros(len(Ts_plot))
for i, T in enumerate(Ts_plot):
    if T <= 2.269:
        Onsager_Ms[i] = (1 - (np.sinh(2/T))**(-4))**(1/8)
plt.plot(Ts, Ms, '^', label='simulation')
plt.plot(Ts_plot, Onsager_Ms, '--', label='theoretical')
plt.ylim(0, 1)
plt.legend()
plt.xlabel('Temperature')
plt.ylabel('Magnetization')
```

```python
# 能量
plt.plot(Ts, Es, '^')
plt.xlabel('Temperature')
plt.ylabel('Energy')
```

```python
# 热容
plt.plot(Ts, CVs, '^')
plt.xlabel('Temperature')
plt.ylabel('Heat capacity')
```
